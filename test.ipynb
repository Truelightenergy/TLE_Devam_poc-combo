```python
import pandas as pd
```

```python
df = pd.read_csv('/home/alee/Documents/Techliance/poc-combo/buildContext/good_test_data/ERCOT_AncillaryData_03162023_090000.csv', skiprows=2)
df.drop(index=[0,1,2], inplace=True, axis='rows')
df
```

```python
# fill missing values with specified values for each column
fill_values = {
    col: '$0' if '$' in str(df[col]) else '0%'
    for col in df.columns
}
df.fillna(fill_values, inplace=True)
df[df.columns[1:]] = df[df.columns[1:]].replace('[\$,%]', '', regex=True).astype(float)
# date parsing
df['Zone'] = pd.to_datetime(df['Zone'])
```

```python
df = df.rename(columns={
    'Zone':'month',

    'RRS Future':'rrs_future',
    'NS Future':'ns_future',
    'RU Future':'ru_future', 
    'RD Future':'rd_future', 
    'RD PHYS':'rd_phys',
    'NS PHYS':'ns_phys',
    'RU PHYS':'ru_phys',
    'RRS PHYS':'rrs_phys'
})
```
